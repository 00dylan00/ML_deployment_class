## 2. Seleccionar un modelo específico: Clasificación de Texto
En este caso, seleccionamos 'distilbert-base-uncased-finetuned-sst-2-english'

```python
selected_model = "distilbert-base-uncased-finetuned-sst-2-english"
```

## 3. Correr el modelo
Crear un pipeline de clasificación de texto

```python
classifier = pipeline("text-classification", model=selected_model)
```

Probar el modelo con algunos textos

```python
text_english = [
    "Patient1: This medication significantly reduced my pain and improved my mobility.",
    "Patient2: The side effects of this drug were severe and difficult to manage.",
    "Patient3: The treatment was somewhat effective, but I still experience mild symptoms.",
    "Patient4: I suffer from constant pain, but have noticed improvements with the treatment, I would like stronger treatment"
]

texto_castellano = [
    "Paciente1: Este medicamento redujo significativamente mi dolor y mejoró mi movilidad.",
    "Paciente2: Los efectos secundarios de este medicamento fueron graves y difíciles de manejar.",
    "Paciente3: El tratamiento fue algo efectivo, pero todavía experimento síntomas leves.",
    "Paciente4: Sufro de dolor constate, pero he notado mejoría con el tratamiento, me gustaría tratamiento más fuerte"]
```

Mostrar resultados

```python
results = classifier(text_english)

for text, result in zip(texto_castellano, results):
    print(f"Texto: {text}\nPredicción: {result['label']} ({result['score']:.2f})\n")
```

```python
import gradio as gr
from transformers import pipeline

# Inicializar los pipelines
clasificador_texto = pipeline(
    "text-classification",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

generador_respuesta_medica = pipeline(
    "text-generation",
    model="ContactDoctor/Bio-Medical-Llama-3-2-1B-CoT-012025",
    max_length=1000,
    temperature=0.7,  # Ajustar para controlar la aleatoriedad
    top_p=0.9         # Mejorar la coherencia
)

# Definir la función para generar opiniones médicas
def generar_opinion_medica(entrada_usuario):
    """
    Genera una opinión médica profesional basada en la descripción proporcionada por el usuario.

    Args:
        entrada_usuario (str): Descripción de los síntomas o escenario médico proporcionado por el usuario.

    Returns:
        str: Respuesta generada profesionalmente por el modelo.
    """
    # Preparar la entrada para el pipeline de generación de texto
    entrada_formateada = [
        {"role": "user", "content": entrada_usuario}
    ]

    # Generar la respuesta usando el pipeline de generación de texto médico
    try:
        respuesta = generador_respuesta_medica(entrada_usuario)
        texto_generado = respuesta[0]["generated_text"]  # Extraer el texto generado
        return texto_generado
    except Exception as e:
        return f"Ocurrió un error al generar la respuesta: {e}"

# Crear un estilo CSS personalizado para mejorar la estética
css_personalizado = """
body {
    font-family: 'Arial', sans-serif;
    background-color: #f9f9f9;
    margin: 0;
    padding: 0;
}
.gradio-container {
    border-radius: 10px;
    border: 1px solid #ccc;
    box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1);
    background-color: white;
}
.header {
    text-align: center;
    margin-bottom: 20px;
}
.header img {
    max-width: 150px;
    margin-bottom: 10px;
}
.header h1 {
    color: #333;
    margin: 0;
}
.header p {
    color: #666;
    margin: 5px 0 15px;
}
"""

# Crear y lanzar la interfaz de Gradio
interfaz = gr.Interface(
    fn=generar_opinion_medica,
    inputs=[
        gr.Textbox(
            lines=5,
            placeholder="Describe los síntomas o el escenario médico (por ejemplo, 'Un hombre de 45 años con dolor crónico en el pecho...')",
            label="Describe el Escenario Médico"
        )
    ],
    outputs=gr.Textbox(
        lines=10,
        placeholder="La respuesta médica generada aparecerá aquí.",
        label="Opinión Médica Generada"
    ),
    title="Generador de Opiniones Médicas",
    description=(
        "### Asistencia Médica Potenciada por IA\n"
        "Esta herramienta genera opiniones médicas profesionales basadas en descripciones proporcionadas de síntomas o escenarios médicos. \n"
        "Se apoya en modelos avanzados de procesamiento del lenguaje natural para proporcionar respuestas coherentes y significativas. Por favor, ten en cuenta que esta herramienta es solo para fines informativos y no reemplaza el consejo médico profesional."
    ),
    article=(
        "#### Descargo de responsabilidad\n"
        "Esta herramienta de IA está diseñada para asistir en escenarios relacionados con la medicina, pero no tiene como objetivo proporcionar diagnósticos definitivos ni planes de tratamiento. Siempre consulta a un profesional médico calificado para obtener asesoramiento médico."
    ),
    examples=[
        ["Un hombre de 45 años con antecedentes de tabaquismo presenta dolor crónico en el pecho y dificultad para respirar. Informa fatiga y tos persistente."],
        ["Una mujer de 30 años con dolor abdominal intermitente y náuseas. Los síntomas comenzaron hace dos días después de consumir mariscos."]
    ],
    theme="compact",
    live=True,
)

# Insertar el CSS para mejorar la estética de la interfaz
def agregar_css_personalizado(css):
    with open("gradio.css", "w") as archivo_css:
        archivo_css.write(css)

agregar_css_personalizado(css_personalizado)
interfaz.launch()
```

## 2. Detector de imagenes

```python
# Importara librerías
from transformers import pipeline
import gradio as gr
from PIL import Image

# variables
tarea = "image-classification"
nombre_modelo = "google/vit-base-patch16-224"

modelo_imagenes = pipeline(tarea, model=nombre_modelo, use_fast=True)

# Funciones
def clasificar_imagen(imagen):
    """
    Clasifica una imagen utilizando el modelo ViT de Hugging Face.

    Args:
        imagen (PIL.Image): Imagen cargada desde Gradio.

    Returns:
        dict: Lista de predicciones con etiquetas y puntajes.
    """
    try:
        # Debug: Check the input type
        print("DEBUG: Tipo de imagen cargada:", type(imagen))
        if imagen is None:
            return "Error: No se cargó ninguna imagen."

        # Debug: Check image details
        print("DEBUG: Detalles de la imagen:", imagen.size, imagen.format)

        # Make predictions
        predicciones = modelo_imagenes(imagen)

        # Debug: Check the predictions
        print("DEBUG: Predicciones generadas:", predicciones)

        # Return predictions in the required format
        resultado = {pred["label"]: pred["score"] for pred in predicciones}
        return resultado
    except Exception as e:
        # Log the error details for debugging
        print(f"ERROR: Ocurrió un error durante la predicción: {e}")
        return f"Error durante la predicción: {e}"



# Create the Gradio interface
interfaz = gr.Interface(
    fn=clasificar_imagen,  # Function to call
    inputs=gr.Image(type="pil", label="Carga tu imagen"),  # Input type: PIL Image
    outputs=gr.Label(num_top_classes=5, label="Predicciones del Modelo"),  # Output: Top 5 predictions
    title="Clasificador de Imágenes con Hugging Face",
    description=(
        "Esta aplicación utiliza el modelo preentrenado ViT (Vision Transformer) para clasificar imágenes. "
        "Carga una imagen para obtener las categorías más probables."
    ),
    examples=[
        ["https://upload.wikimedia.org/wikipedia/commons/9/9a/Pug_600.jpg"],  # Example: Pug
        ["https://upload.wikimedia.org/wikipedia/commons/3/3a/Cat03.jpg"],   # Example: Cat
    ],
    live=True,  # Enable live prediction
)

# Launch the interface
interfaz.launch(debug=True)
```
